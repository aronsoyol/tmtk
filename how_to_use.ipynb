# TMTK 使用方法

## 切词

```python
from tmtk import tokenizer

text = ("""ᠪᠢᠳᠡ ᠪᠣᠯ ᠬᠥᠮᠦᠨ ᠲᠥᠷᠦᠯᠬᠢᠲᠡᠨ ᠦ᠌ ᠰᠢᠨ᠎ᠡ ᠲᠡᠦᠬᠡ ᠶ᠋ᠢ ᠨᠡᠭᠡᠭᠡᠵᠦ ᠪᠠᠢᠭ᠎ᠠ """
"""ᠭᠡᠷᠡᠯ ᠰᠠᠴᠤᠷᠠᠩᠭᠤᠢ ᠴᠠᠭ ᠦᠶ᠎ᠡ ᠳ᠋ᠦ ᠠᠮᠢᠳᠤᠷᠠᠵᠤ ᠪᠠᠢᠭ᠎ᠠ ᠶᠤᠮ᠃ """
"""ᠡᠨᠡ ᠬᠦ ᠴᠠᠭ ᠦᠶ᠎ᠡ ᠳ᠋ᠦ ᠮᠠᠰᠢ ᠣᠯᠠᠨ ᠬᠡᠷᠡᠭ᠌ ᠪᠣᠳᠠᠰ ᠲᠤ ᠬᠥᠮᠦᠰ """
"""ᠰᠢᠨ᠎ᠡ ᠪᠣᠳᠤᠯ ᠰᠢᠨ᠎ᠡ ᠲᠥᠷᠦᠭᠳᠡᠯ ᠣᠷᠤᠵᠤ ᠪᠠᠢᠨ᠎ᠠ᠃""")

tokens = [a for a in tokenizer.tokenize(text,split_suffix=True,only_mongolian=True)]
    
print(tokens)
```
